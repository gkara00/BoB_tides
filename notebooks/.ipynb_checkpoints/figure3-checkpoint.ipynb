```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams["font.size"] = 14
```

```python
def read_netcdf_xr(fid):
    ds = xr.open_mfdataset(fid, combine='by_coords')
    return ds

root_dir = '/home/gkara/Documents/for-tide_paper/'
_f = ['METOFFICE-GLO-SST-L4-NRT-OBS-SST-V2_20110928_.nc', 'METOFFICE-GLO-SST-L4-NRT-OBS-SST-V2_20120224_.nc']

###########
T_summer = read_netcdf_xr(root_dir + _f[0])
#
T_winter = read_netcdf_xr(root_dir + _f[1])
# Convert the dataset from Kelvin to degrees Celsius 
T_summer = T_summer - 273
T_winter = T_winter - 273
# land mask
land_mask = (T_summer.analysed_sst > 0)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5,7))
for i in range(2):
    if i==0:
        vmin = 14
        vmax = 22
        levels = 33
        cbar_kwargs = {"label": 'Temperature ($^\circ C$)', "ticks":np.arange(vmin,vmax+1), "pad":0.02}
        T_summer.isel(time=0).analysed_sst.plot.contourf(
            ax=axes[i], x='lon', y='lat', levels=levels, vmin=vmin, vmax=vmax, \
            xlim=(-11,0), ylim=(41,52), cmap='RdYlBu_r', \
            add_colorbar=True, extend='both', cbar_kwargs=cbar_kwargs)
        land_mask[0,:,:].plot.contour(ax=axes[i], linewidths=0.2, colors='k')
    else:
        vmin = 8
        vmax = 14
        levels = 25
        cbar_kwargs = {"label": 'Temperature ($^\circ C$)', "ticks":np.arange(vmin,vmax+1), "pad":0.02}
        T_winter.isel(time=0).analysed_sst.plot.contourf(
            ax=axes[i], x='lon', y='lat', levels=levels, vmin=vmin, vmax=vmax, \
            xlim=(-11,0), ylim=(41,52), cmap='RdYlBu_r', \
            add_colorbar=True, extend='both', cbar_kwargs=cbar_kwargs)
        land_mask[0,:,:].plot.contour(ax=axes[i], linewidths=0.2, colors='k')

# annotate subplots
idx = [('$(a)$','28 Sep. 2011'), ('$(b)$', '24 Feb. 2012')]
for i, ax in enumerate(axes.flat):
    ax.axes.axis('tight')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title('')
    ax.text(-1.2, 41.8, idx[i][0], size=14, weight='bold')
    ax.text(-7, 41.8, idx[i][1], size=14, weight='bold')

# set figure title
axes[0].set_title('Satellite SST data')
plt.tight_layout()

#save figure
fig.savefig('../_plots/figure3.png',dpi=300,bbox_inches='tight')
```
