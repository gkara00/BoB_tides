```python
import xarray as xr
import matplotlib.pyplot as plt
import cmaps
import numpy as np
import cmocean

%matplotlib inline
plt.rcParams["font.size"] = 14
```

```python
def read_netcdf_xr(fid):
    ds = xr.open_mfdataset(fid, combine='by_coords')
    return ds
```

```python
root_dir = '/home/gkara/Documents/for-tide_paper/'
_f = ['BISCAY-T60_1d_gridT_20110928-20110928.nc', 'BISCAY-T60_1d_gridT_20120224-20120224.nc']

# load T files -- TON (3D)
T_ref = read_netcdf_xr([root_dir + 'ref/' + f for f in _f])
T_ref = T_ref.set_coords(['nav_lon', 'nav_lat'])

# load T files --TOFF (3D)
T_exp1 = read_netcdf_xr([root_dir + 'exp1/' + f for f in _f])
T_exp1 = T_exp1.set_coords(['nav_lon', 'nav_lat'])

# load bathymetry
bathy = read_netcdf_xr(root_dir + 'bathy_meter.nc')
bathy = bathy.set_coords(['nav_lon', 'nav_lat'])
land_mask = (bathy.Bathymetry > 0)

# mask the land
T_ref = T_ref.where(land_mask)
T_exp1 = T_exp1.where(land_mask)
```

```python
fton = ['TON_bottom_gridT_20110928.nc']
ftoff = ['TOFF_bottom_gridT_20110928.nc']

# load T files -- TON (bottom level)
T_refb = xr.open_mfdataset([root_dir + f for f in fton], combine='by_coords')
T_refb = T_refb.set_coords(['nav_lon', 'nav_lat'])

# load T files --TOFF (bottom level)
T_exp1b = xr.open_mfdataset([root_dir + f for f in ftoff], combine='by_coords')
T_exp1b = T_exp1b.set_coords(['nav_lon', 'nav_lat'])

# mask the land
T_refb = T_refb.where(land_mask)
T_exp1b = T_exp1b.where(land_mask)
```

```python
def simpsonhunter(u, v, depth):
    """
    Calculate the Simpson-Hunter parameter (h/u^{3}).

    Parameters
    ----------
    u, v : ndarray
        Depth-resolved current vectors.
    depth : ndarray
        Water depth (m). Must be on the same grid as u and v.

    Returns
    -------
    SH : ndarray
        Simpson-Hunter parameter (np.log10(m^{-2}s^{-3})).

    References
    ----------
    - Simpson, JH, and JR Hunter. "Fronts in the Irish Sea." Nature 250 (1974):
      404-6.
    - Holt, Jason, and Lars Umlauf. "Modelling the Tidal Mixing Fronts and
      Seasonal Stratification of the Northwest European Continental Shelf."
      Continental Shelf Research 28, no. 7 (April 2008): 887-903.
      doi:10.1016/j.csr.2008.01.012.

    """
    
    # calculate speed
    uv = np.sqrt(u**2 + v**2)

    # Take the average of the maxima for the parameter calculation.
    uvbar = uv.mean(axis=0)

    SH = np.log10(depth / np.sqrt((uvbar**3)**2))

    return SH

from pathlib import Path

data_path = Path("/home/gkara/Documents/for-tide_paper/")

u_summer = "*1h_gridU_20110[7-9]*"
v_summer = "*1h_gridV_20110[7-9]*"
fsummer_u = (list(sorted(data_path.glob('ref/' + u_summer))))
fsummer_v = (list(sorted(data_path.glob('ref/' + v_summer))))

# load data
u_summer = read_netcdf_xr(fsummer_u)
v_summer =  read_netcdf_xr(fsummer_v)

# mask land
u_summer = u_summer.where(land_mask)
v_summer = v_summer.where(land_mask)

# select dataset arrays for u/v
u = u_summer.sozobrtx.values
v = v_summer.somebrty.values

# Calculate SH parameter
SH = simpsonhunter(u, v, bathy.Bathymetry.values)
```
