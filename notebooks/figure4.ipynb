```python
import xarray as xr
import matplotlib.pyplot as plt
import cmaps
import numpy as np

%matplotlib inline
plt.rcParams["font.size"] = 14
```

```python
def read_netcdf_xr(fid):
    ds = xr.open_mfdataset(fid, combine='by_coords')
    return ds

def plot_200_mask(axes, mask, bathy):
    '''function for plotting 200m isobath'''
    for ax in axes.flat:
        mask.plot.contour(ax=ax, x='nav_lon', y='nav_lat', linewidths=0.1, colors='k')
        bathy.Bathymetry.plot.contour(ax=ax, x='nav_lon', y='nav_lat', levels=[200], linewidths=1.1, colors='k')
    return
```

```python
root_dir = '/home/gkara/Documents/for-tide_paper/'
_f = ['BISCAY-T60_1d_gridT_20110928-20110928.nc', 'BISCAY-T60_1d_gridT_20120224-20120224.nc']

# load T files -- TON
T_ref = read_netcdf_xr([root_dir + 'ref/' + f for f in _f])
T_ref = T_ref.set_coords(['nav_lon', 'nav_lat'])

# load T files --TOFF
T_exp1 = read_netcdf_xr([root_dir + 'exp1/' + f for f in _f])
T_exp1 = T_exp1.set_coords(['nav_lon', 'nav_lat'])

# load bathymetry
bathy = read_netcdf_xr(root_dir + 'bathy_meter.nc')
bathy = bathy.set_coords(['nav_lon', 'nav_lat'])
land_mask = (bathy.Bathymetry > 0)

# select surface level and mask land
T_ref = T_ref.isel(deptht=0)
T_exp1 = T_exp1.isel(deptht=0)
T_ref = T_ref.where(land_mask)
T_exp1 = T_exp1.where(land_mask)
```

```python
# compute difference: TON - TOFF
diff = T_ref - T_exp1
mask = np.isnan(T_ref.votemper[0,:,:])

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16,9))
for i in range(2):
    
    if i==0:
        vmin = 14
        vmax = 22
        levels = 33 #43
        cbar_kwargs = {"label": 'Temperature ($^\circ C$)', "ticks":np.arange(vmin,vmax+1), "pad":0.02}
    else:
        vmin = 8
        vmax = 14
        levels = 25#31
        cbar_kwargs = {"label": 'Temperature ($^\circ C$)', "ticks":np.arange(vmin,vmax+1), "pad":0.02}
        
    T_ref.votemper.isel(time_counter=i).plot.contourf(
        ax=axes[i, 0], x='nav_lon', y='nav_lat', levels=levels, vmin=vmin, vmax=vmax, cmap='RdYlBu_r',
        add_colorbar=True, extend='both', cbar_kwargs=cbar_kwargs)
    T_exp1.votemper.isel(time_counter=i).plot.contourf(
        ax=axes[i, 1], x='nav_lon', y='nav_lat', levels=levels, vmin=vmin, vmax=vmax, cmap='RdYlBu_r',
        add_colorbar=True, extend='both', cbar_kwargs=cbar_kwargs)

    diff.votemper.isel(time_counter=i).plot.contourf(
        ax=axes[i, 2], x='nav_lon', y='nav_lat', levels=9, vmin=-1, vmax=1, cmap='RdBu_r',
        add_colorbar=True, extend='both', cbar_kwargs={"label": 'Temperature ($^\circ C$)', "pad":0.02})

# plot 200 m isobath
plot_200_mask(axes=axes, mask=mask, bathy=bathy)    

# annotate subplots
idx = ['$(a)$', '$(b)$', '$(c)$', '$(d)$', '$(e)$', '$(f)$']
for i, ax in enumerate(axes.flat):
    ax.axes.axis('tight')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title('')
    ax.text(-1, 41.5, idx[i], size=14, weight='bold')

# set titles
axes[0, 0].set_title('TON')
axes[0, 1].set_title('TOFF')
axes[0, 2].set_title('Difference: TON-TOFF')

plt.tight_layout()

#save figure
fig.savefig('../_plots/figure4.png',dpi=300,bbox_inches='tight')
```
